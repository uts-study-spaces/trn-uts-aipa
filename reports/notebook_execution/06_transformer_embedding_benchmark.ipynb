# **06. Optional Multilingual Transformer-Embedding Benchmark**

This notebook adds a time-permitting benchmark for AT3. It compares the current TF-IDF supervised NLP pipeline with a multilingual transformer embedding representation followed by a simple supervised classifier. It is not required for the Streamlit demo and does not replace the final Linear SVM app model.

# **1. Environment Setup**

```python
# Import libraries and project helper functions.
from pathlib import Path

import pandas as pd

from customer_support_ai.config import RESULTS_DIR
from customer_support_ai.transformer_benchmark import (
    DEFAULT_EMBEDDING_MODEL,
    OUTPUT_CSV,
    run_benchmark,
)

print('Embedding model:', DEFAULT_EMBEDDING_MODEL)
print('Benchmark output:', OUTPUT_CSV)
```

# **2. Benchmark Purpose**

The baseline and comparison models in the main pipeline all use the same main paradigm: supervised NLP text classification with TF-IDF features. This optional benchmark adds a different representation approach: dense semantic multilingual transformer embeddings. The classifier is still supervised, but the language representation is transformer-based rather than sparse lexical TF-IDF.

# **3. Run Sampled Transformer Benchmark**

The benchmark uses a small language-balanced sample by default so it can run on a laptop. Increase `sample_size` only if time and compute allow it. Results should be compared cautiously against the full-data TF-IDF results because this is a sampled experiment.

```python
# Run the optional benchmark. Install optional dependencies first if needed:
#   python -m pip install -r requirements-transformer.txt

benchmark_results = run_benchmark(sample_size=2500)
benchmark_results
```

# **4. Compare With Current TF-IDF Results**

```python
# Load the current full-data TF-IDF model results and compare headline macro F1 values.
tfidf_results = pd.read_csv(RESULTS_DIR / 'metrics_summary.csv')
tfidf_test = tfidf_results[tfidf_results['split'] == 'test'][['task', 'model', 'split', 'macro_f1', 'accuracy']]

comparison = pd.concat(
    [
        tfidf_test.assign(source='Full-data TF-IDF pipeline'),
        benchmark_results[['task', 'model', 'split', 'macro_f1', 'accuracy']].assign(source='Sampled transformer benchmark'),
    ],
    ignore_index=True,
)
comparison.sort_values(['task', 'source', 'macro_f1'], ascending=[True, True, False])
```
